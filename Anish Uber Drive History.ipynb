# Analyzing Data
1) Analyzing by Weekday

2)Total number of Uber pickups on each day


3)creating bar plot using matplotlib 

```python
week_day=pd.DataFrame(df['weekday'].value_counts())
week_day
```

```python
week_day.plot.bar()
plt.title('Total Uber pickups')
plt.xlabel('Weekday')
plt.ylabel('Frequency')
plt.show()
```

#note: I am considering Friday and saturday as a weekend because FRIDAY is the busiest day.

```python
week_day['day_type']=['we','we','wd','wd','wd','wd','wd']
```

```python
week_day
```

```python
#Weekends vs Weekdays : Difference in average number of pickups on weekdays and weekends
pie=week_day.groupby('day_type').sum()
pie.iloc[0,0]=pie.iloc[0,0]/5
pie.iloc[1,0]=pie.iloc[1,0]/2
pie
```

Lets makes pie-chart to visualized 

```python
explode=[0,0.05]
colors=['red','blue']
labels=['Weekday','Weekend']
pie.plot.pie(autopct = '%1.1f%%',shadow=True,subplots=True,
        colors=colors,explode=explode)
plt.legend(labels=labels)
plt.show()
```

4) Hour-wise Analysis



```python
#first we add new column and see hourly wise analysis


df['hour']=df['Local Request Timestamp'].dt.hour
hour=df.groupby('hour').count()
df
```

```python
hour=pd.DataFrame(hour['Begin Trip Latitude'])
hour.columns=['Frequency']
hour
```

```python
hour.plot.bar()
plt.ylabel('Frequency')
plt.title('Uber pickups by hour')
plt.show()
```

Uber pickups tend to be maximum around 5-6 pm, when the work-day for most poeple ends. This trend can vary for weekends, thus separately checking for weekdays and weekends

For Weekends:


```python
weekend = df[(df['weekday']=='Sunday') | (df['weekday']=='Saturday')]
```

```python
weekend=weekend.groupby('hour').count()
weekend=pd.DataFrame(weekend['Begin Trip Latitude'])
weekend.columns=['Frequency']
```

```python
weekend
```

```python
weekend.plot.bar()
plt.ylabel('Frequency')
plt.title('Uber pickups by hour - Weekends')
plt.show()
```

For Weekdays:

```python
weekday= df [(df['weekday']=='Monday') | (df['weekday']=='Tuesday') | (df['weekday']=='wednesday')| (df['weekday']=='Thursday')| (df['weekday']=='Friday')]
```

```python
weekday=weekday.groupby('hour').count()
weekday=pd.DataFrame(weekday['Begin Trip Latitude'])
weekday.columns=['Frequency']
```

```python
#lets add Date of month and Month as new column serperately
def get_dom(dt):
    return dt.day

df['Date_of Month'] = df['Local Request Timestamp'].map(get_dom)
```

```python
def get_moy(dt):
    return dt.month

df['Month'] = df['Local Request Timestamp'].map(get_moy)
```

```python
%pylab inline
hist(df.Month, bins=12, rwidth=0.9, color="teal", range=(0.1, 12))
xlabel('Month')
ylabel('frequency')
title('Frequency by Month')
```

we got to know that july,August and september are the top 3 busiest moth for me


```python
#trying to find the correlation between hour and weekday( got an idea to find busiest during weekday)
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for weekday and hours")

plt.xlabel("weekday")

plt.ylabel("hour")

plt.show()
```
